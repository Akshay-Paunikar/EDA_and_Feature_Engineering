```python
# read the country code dataset
df_country = pd.read_excel('E:\iNeuron\EDA_and_Feature_Engineering\Day_1_Zomato_Dataset\Zomatodataset\Country-Code.xlsx')

# check records
df_country.head(20)
```

```python
# combine the 2 dataframes
final_df = pd.merge(df, df_country, on='Country Code',how='left')

# check the records
final_df.head()
```

```python
# check for missing values
final_df.isnull().sum()
```

```python
# check for duplicated records
final_df.duplicated().sum()
```

```python
# Find out the top 3 countries who are using the zomato app
final_df.Country.value_counts()
```

```python
country_names = final_df['Country'].value_counts().index
country_values = final_df['Country'].value_counts().values
```

```python
# create a pie chart showing top 3 countries
plt.figure(figsize=(6,6))
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.title("Top 3 Countries using Zomato")
plt.show()
```

#### The top 3 countries that are using zomato are : India - 94.39%, United States - 4.73%, United Kingdom - 0.87%

```python
# Create a separare data for ratings
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

#### INSIGHTS from above table
 - Rating below 2.5 --->>> "Poor"
 - Rating between 2.5 - 3.4 --->>> "Average"
 - Rating between 3.5 - 3.9 --->>> "Good"
 - Rating between 4.0 - 4.4 --->>> "Very Good"
 - Rating above 4.4 --->>> "Excellent"

```python
# Let's visualize above data using barplot
plt.figure(figsize=(12, 4))
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color',data=ratings, 
            palette=['blue', 'red', 'orange', 'yellow', 'lightgreen', 'green'],dodge=False)
plt.xlabel("Rating")
plt.ylabel("Counts")
plt.title("Aggregate Rating Counts")
plt.show()
```

#### Maximum number of ratings are between 2.6 to 4.2

```python
# Countplot for Ratings Color
plt.figure(figsize=(8, 4))
sns.barplot(x='Rating color',y='Rating Count', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'lightgreen', 'green'],
                estimator=np.sum,dodge=False, ci=None)
plt.xlabel("Rating Color")
plt.ylabel("Counts")
plt.title("Countplot for Ratings Color")
plt.show()
```

```python
# Find the countries that given 0 ratings or not rated
zero_rated_country = final_df[final_df['Rating color']=='White'].groupby(['Country']).size().reset_index().rename(columns={0:"Counts"})
zero_rated_country
```

```python
zero_rated_country.values[:,1]
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Country',y='Counts', data=zero_rated_country, dodge=False, ci=None)
plt.xlabel("Country")
plt.ylabel("Counts")
plt.title("Countries who gave 0 ratings/not rated")
plt.show()
```

#### Maximum number of 0 rating/not rated is given by Indian Customers.

```python
# Country and the currency used.
final_df[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Top 5 Cities
final_df['City'].value_counts().head()
```

```python
city_names = final_df['City'].value_counts().index
city_values = final_df['City'].value_counts().values
```

```python
# create a pie chart showing top 3 countries
plt.figure(figsize=(6,6))
plt.pie(city_values[:5], labels=city_names[:5], autopct='%1.2f%%')
plt.title("Top 5 Cities using Zomato")
plt.show()
```

#### The top 5 cities that are using Zomato are: New Delhi - 68.87%, Gurgaon - 14.07%, Noida - 13.59%, Faridabad - 3.16%, Ghaziabad - 0.31%.

```python
# Let's find out the top restaurants in India according to ratings
top_5_rest_india = final_df[final_df['Country']=='India'].groupby(['Restaurant Name', 'City'])['Aggregate rating'].mean().sort_values(
    ascending=False).reset_index().head()
top_5_rest_india
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y='Restaurant Name', x='Aggregate rating', hue='City', data=top_5_rest_india, dodge=False)
plt.title("Top 5 restaurants in India by Aggregate Rating")
plt.show()
```

#### Above barplot shows the top 5 restaurants in India with the city according to Aggregate Ratings.

```python
# Top 10 Cuisines in the world
top_cuisines = final_df['Cuisines'].value_counts().reset_index().rename(columns={'index':'Cuisines', 'Cuisines':'Counts'}).head(10)
top_cuisines
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y='Cuisines', x='Counts', data=top_cuisines, dodge=False)
plt.title("Top 10 Cuisines in the world")
plt.show()
```

#### North Indian, Chinese, Fast Food are some of the top Cuisines in the data.

```python
# Top 10 Restaurant by Votes
top_rest_votes = final_df.groupby(['Restaurant Name', 'City'])['Votes'].mean().sort_values(ascending=False).reset_index().head(10)
top_rest_votes
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y='Restaurant Name', x='Votes', hue='City', data=top_rest_votes, dodge=False)
plt.title("Top 10 Restaurant by Votes")
plt.show()
```

#### Restaurants from Bangalore has maximum votes

```python
# Top 10 restaurants according to average cost for two
top_avg_cost = final_df.groupby(['Restaurant Name', 'City'])['Average Cost for two'].mean().sort_values(
    ascending=False).reset_index().head(10)
top_avg_cost
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y='Restaurant Name', x='Average Cost for two', hue='City', data=top_avg_cost, dodge=False)
plt.title("Top 10 restaurants according to average cost for two")
plt.show()
```

#### Above barplot shows Top 10 restaurants according to average cost for two

```python
# Bottom 10 restaurants according to average cost for two
bottom_avg_cost = final_df[final_df['Average Cost for two']>0.0].groupby(['Restaurant Name', 'City'])['Average Cost for two'].mean().sort_values(
    ascending=True).reset_index().head(10)
bottom_avg_cost
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y='Restaurant Name', x='Average Cost for two', hue='City', data=bottom_avg_cost, dodge=False)
plt.title("Bottom 10 restaurants according to average cost for two")
plt.show()
```

#### Above barplot shows Bottom 10 restaurants according to average cost for two

```python
# Top 10 restaurants in India by Average Cost for two
top_india_avg_cost = india[india['Average Cost for two']>0.0].groupby(['Restaurant Name'])['Average Cost for two'].mean().sort_values(
    ascending=False).reset_index().head(10)
top_india_avg_cost
```
