```python
# Find out the top 3 countries who are using the zomato app
final_df.Country.value_counts()
```

```python
country_names = final_df['Country'].value_counts().index
country_values = final_df['Country'].value_counts().values
```

```python
# create a pie chart showing top 3 countries
plt.figure(figsize=(6,6))
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.title("Top 3 Countries using Zomato")
plt.show()
```

#### The top 3 countries that are using zomato are : India - 94.39%, United States - 4.73%, United Kingdom - 0.87%

```python
# Create a separare data for ratings
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

#### INSIGHTS from above table
 - Rating below 2.5 --->>> "Poor"
 - Rating between 2.5 - 3.4 --->>> "Average"
 - Rating between 3.5 - 3.9 --->>> "Good"
 - Rating between 4.0 - 4.4 --->>> "Very Good"
 - Rating above 4.4 --->>> "Excellent"

```python
# Let's visualize above data using barplot
plt.figure(figsize=(12, 4))
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color',data=ratings, 
            palette=['blue', 'red', 'orange', 'yellow', 'lightgreen', 'green'],dodge=False)
plt.xlabel("Rating")
plt.ylabel("Counts")
plt.title("Aggregate Rating Counts")
plt.show()
```

#### Maximum number of ratings are between 2.6 to 4.2

```python
# Countplot for Ratings Color
plt.figure(figsize=(8, 4))
sns.barplot(x='Rating color',y='Rating Count', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'lightgreen', 'green'],
                estimator=np.sum,dodge=False, ci=None)
plt.xlabel("Rating Color")
plt.ylabel("Counts")
plt.title("Countplot for Ratings Color")
plt.show()
```

```python
# Find the countries that given 0 ratings or not rated
zero_rated_country = final_df[final_df['Rating color']=='White'].groupby(['Country']).size().reset_index().rename(columns={0:"Counts"})
zero_rated_country
```

```python
zero_rated_country.values[:,1]
```
